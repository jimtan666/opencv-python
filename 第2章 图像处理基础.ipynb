```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

# 2.1 图像的基本表示方法

### 2.1.1 二值图像

```python
# 白色像素点存储为1
# 黑色像素点存储为0
# 1个比特就能表示[0,1]
```

### 2.1.2 灰度图像

```python
# 灰度值：[0,255]
# 一个字节表示
# 也可用灰度图表示二值图：像素值只有0和255
```

### 2.1.3 彩色图像

```python
# RGB时其中一种色彩空间
# RGB三通道，每个通道像素值[0,255]
# 三个通道组合表示颜色
# 注意：opencv中通道的顺序时BGR
```

# 2.2 像素处理

### 2.2.1 二值图及灰度图

```python
# opencv中最小数据类型时无符号8位数，因此不能表示二值图，经过处理，0表示黑，255表示白，将灰度图处理成二值图，二值图时特殊的灰度图
# 可以通过位置访问图像内元素，图像用Numpy.array格式存储
```

```python
# Numpy生成元素全0数组，模拟黑色图像，对元素进行访问，修改（下标访问修改，先行后列）

img = np.zeros((8,8), dtype=np.uint8)
print("imgArray:\n", img)

cv2.imshow('one', img)
print("读取像素点img[0,3]:", img[0,3])

img[0,3] = 255
print("修改后的img:\n", img)

cv2.imshow("two:", img)

# cv2.waitKey()
# cv2.destroyAllWindows()
```

```python
img
```

```python
# 补充：用cv2.waitKey()，cv2.destroyAllWindows()显示销毁图片的窗口关闭不了，可用matplotlib代替显示图像
img = cv2.imread("/Users/tanjun/Desktop/加班截图/WechatIMG115.jpeg")

def img_show(img):
    b,g,r = cv2.split(img)     # 分割出3个通道
    img2 = cv2.merge([r,g,b])  # 合并成新图片
    plt.imshow(img2)           # 显示图片
    
img_show(img)
```

### 2.2.1 彩色图像

```python
# opencv读取图像按BGR通道顺序读入像素点，并将像素以行为单位存储在ndarray的列中
# 可使用img[0,0,0]访问B通道第0行第0列像素点（img[行，列，通道]）

# 生成数组，观察3通道情况
# ---------蓝色通道----------
blue = np.zeros((300,300,3), dtype=np.uint8)
blue[:,:,0] = 255
img_show(blue)
# ---------绿色通道----------
gree = np.zeros((300,300,3), dtype=np.uint8)
gree[:,:,1] = 255
img_show(gree)
# ---------红色通道----------
red = np.zeros((300,300,3), dtype=np.uint8)
red[:,:,2] = 255
img_show(red)
```

```python
red
```

```python
# 生成一个三位数组，观察3通道
img = np.zeros((300,300,3), dtype=np.uint8)

img[:,0:100,0] = 255
img[:,100:200,1] = 255
img[:,200:300,2] = 255

img_show(img)
```

# 2.3 使用numpy.array访问像素

```python
# numpy.array提供了item()和itemset()函数访问像素，这两函数经过了优化，能大幅提高效率，比直接用索引更快
# item(行，列): 高效访问像素
# itemset(索引值，新值): 修改像素
```

### 2.3.1 二值图像及灰度图像

```python
img = np.random.randint(10,99,size=[5,5],dtype=np.uint8)
print("img:\n",img)
print("读取像素值item(3,2)：", img.item(3,2))

img.itemset((3,2),255)
print("修改img:\n",img)
print("修改像素值item(3,2)：", img.item(3,2))
```

### 2.3.2 彩色图像

```python
# item(行，列，通道)
# itemset(三元祖索引，新值)
```

```python
# 随机生成一幅彩色随机数
img = np.random.randint(0,256,size=[256,256,3],dtype=np.uint8)
img_show(img)
```

```python
print(img.item(0,0,0))
img.itemset((0,0,0), 255)
print(img.item(0,0,0))
```

# 2.4 感兴趣区域（ROI）

```python
# 设定感兴趣区域后，就可对该区域整体操作：如将感兴趣区域A赋值给变量B，再将变量B赋值给另一个区域C，达到区域C中赋值区域A的目的
img = cv2.imread("/Users/tanjun/Desktop/加班截图/WechatIMG115.jpeg")
a = img[0:100,0:400]
img[500:600,0:400] = a
img_show(img)
```

# 2.5 通道操作

### 2.5.1 通道拆分

```python
# 1.通过索引拆分
img = cv2.imread("/Users/tanjun/Desktop/加班截图/WechatIMG115.jpeg")

b = img[:,:,0]
g = img[:,:,1]
r = img[:,:,2]

img[:,:,0] = 0
img_show(img)
```

```python
# 2.通过函数拆分：cv2.split(img)
img = cv2.imread("/Users/tanjun/Desktop/加班截图/WechatIMG115.jpeg")
b,g,r = cv2.split(img)
```

### 2.5.2 通道合并

```python
bgr = cv2.merge([b,g,r])
img_show(bgr)
```
